```python
# 1. MOUNT GOOGLE DRIVE
# This step allows the notebook to access files stored in MY Google Drive
drive.mount('/content/drive')

# 2. DATA INGESTION FROM MY DRIVE
# Path to the dataset I uploaded to MY Google Drive folder
file_path = '/content/drive/MyDrive/ML_Project/bank-full.csv'
```

Loading the Dataset



```python
# Loading the dataset with the correct separator (;) used in the UCI bank data
try:
    df = pd.read_csv(file_path, sep=';')
    print("--- Success: Dataset Loaded from MY Google Drive ---")
    print(f"Total Records: {df.shape[0]} | Total Features: {df.shape[1]}")
except FileNotFoundError:
    print("Error: The file path is incorrect. Please check MY Drive folder name.")
```

3. DATA CLEANING & EXPLORATION (EDA)

```python
# Checking for null values to ensure MY data is clean before training
print("\nNull Values in MY Dataset:")
print(df.isnull().sum())

# Visualizing the target variable 'y' to see the class imbalance in MY project
plt.figure(figsize=(6,4))
sns.countplot(x='y', data=df, palette='viridis')
plt.title('Target Distribution: Subscription (Yes/No)')
plt.show()
```
